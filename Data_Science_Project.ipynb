```python
#(h)
# Calculate mean, median, standard deviation, and percentiles for numerical columns
numerical_cols = df.select_dtypes(include=np.number)

def stats(x):
    return pd.Series({
        'Mean': x.mean(),
        'Median': x.median(),
        'S': x.std(),
        '20th Percentile': x.quantile(0.20),
        '40th Percentile': x.quantile(0.40),
        '60th Percentile': x.quantile(0.60),
        '80th Percentile': x.quantile(0.80),
        '100th Percentile': x.quantile(1)
    })

print(numerical_cols.apply(stats))
```

```python
#(j)
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'Age' is a significant attribute for analysis
plt.figure(figsize=(8, 6))
sns.boxplot(x='HeartDisease', y='Age', data=df)
plt.title('Boxplot of Age Distribution by Heart Disease')
plt.xlabel('Heart Disease')
plt.ylabel('Age')
plt.show()

# Justification:
#This boxplot illustrates the age distribution of individuals with (represented as 1) and without (represented as 0) heart disease.
#It allows us to compare the median age, variability, and any outliers between the two groups.
#This visualization helps determine whether age is a significant predictor of heart disease.
```
